## Fast Fourier Transform (FFT)

```python
piano_ft = np.fft.fft(piano)
print(piano_ft.shape)
```

```python
magnitude_spectrum_piano = np.abs(piano_ft)
```

```python
def plot_magnitude_spectrum(signal, title, sr):
    ft = np.fft.fft(signal)
    magnitude_spectrum = np.abs(ft)

    #plot magnitude spectrum
    plt.figure(figsize=(18,5))

    frequency = np.linspace(0, sr, len(magnitude_spectrum))
    plt.plot(frequency, magnitude_spectrum)
    plt.xlabel("Frequence (Hz)")
    plt.title(title)

    plt.show()
```

```python
plot_magnitude_spectrum(piano, "Piano in C Major", sr)
```
